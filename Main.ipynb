## Introduction
The purpose of this analysis is to examine the school attendance rate by student group over the years. The data used for this analysis is sourced from the 'School_Attendance_by_Student_Group_and_District__2021-2022.csv' file. The dataset contains information on student groups, student counts, and attendance rates for the years 2019-2020, 2020-2021, and 2021-2022. This analysis provides insights into the school attendance rate by student group over the years. By examining the attendance rates for each student group, we can identify any variations or trends that may exist. The visualization helps us understand the differences in attendance rates among different student groups and how they have changed over time.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'School_Attendance_by_Student_Group_and_District__2021-2022.csv'
data = pd.read_csv(file_path)
```

```python
filtered_data = data[['Student group', '2021-2022 student count - year to date', '2021-2022 attendance rate - year to date', '2020-2021 attendance rate', '2019-2020 attendance rate']]

grouped_data = filtered_data.groupby('Student group').mean()
grouped_data = grouped_data[['2019-2020 attendance rate', '2020-2021 attendance rate', '2021-2022 attendance rate - year to date']]
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
grouped_data.plot(kind='bar', ax=ax)
ax.set_ylabel('Attendance Rate')
ax.set_xlabel('Student Group')
ax.set_title('Attendance Rate by Student Group Over the Years')
ax.legend(loc='lower left')
plt.show()
```

```python
# Define a list of colors for each category
colors = ['red', 'green', 'blue', 'orange', 'purple', 'pink', 'brown', 'gray', 'cyan', 'magenta', 'yellow', 'black', 'lime']

fig, ax = plt.subplots(figsize=(12, 8))

# Iterate over each ethnicity and assign a different color
for i, ethnicity in enumerate(grouped_data.index):
    ax.plot(grouped_data.columns, grouped_data.loc[ethnicity], marker='o', label=ethnicity, color=colors[i])

ax.set_ylabel('Attendance Rate')
ax.set_xlabel('Year')
ax.set_title('Attendance Rate by Student Group Over the Years')
ax.legend(loc='lower left')
plt.show()
```

## Analysis
The following steps were performed for the analysis:

1. **Data Filtering**: The data was filtered to select specific columns, including 'Student group', '2021-2022 student count - year to date', '2021-2022 attendance rate - year to date', '2020-2021 attendance rate', and '2019-2020 attendance rate'. This allowed us to focus on the relevant information for our analysis.

2. **Grouping and Calculation**: The filtered data was grouped by 'Student group' and the mean attendance rates for each group were calculated. This step helped us identify any variations in attendance rates among different student groups over the years.

3. **Visualization**: The results were visualized using a bar plot. The x-axis represents the different student groups, and the y-axis represents the attendance rate. By comparing the attendance rates across different years, we can observe any changes or patterns in attendance behavior.
